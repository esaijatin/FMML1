```python
# list methods
first_list=[1,1,1,2,2,3,4,'a','b','c']
first_list.append('new')
print(first_list)
x=first_list.copy()
print(x)
y=first_list.count(1)
print(y)
first_list.clear()
print(first_list)
```

```python
#tuple methods
first_tuple=(1,2,'abcdef',(1,2,3))
print(len(first_tuple))
print(type(first_tuple))
```
